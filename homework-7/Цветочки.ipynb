```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(-15,15,100)
y = np.linspace(-15,15,100)
X, Y = np.meshgrid(x, y)

#Z = np.sqrt(X ** 2 + Y ** 2) + 3 * np.cos(np.sqrt(X ** 2 + Y ** 2)) + 5
Z = 5 - 3 / np.sqrt(X ** 2 + Y ** 2) +  np.sin(np.sqrt(X ** 2 + Y ** 2))

fig = plt.figure()
ax = plt.axes(projection='3d')

ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap='RdPu')
plt.show()
```

```python
b = np.linspace(-10,10,1000)
a = np.random.randint(1,10,5)

x = np.sin(b) + a[0] * np.sin(b) + a[1] * np.sin(b * a[2]) + a[3] * np.sin(b * a[4]);
y = np.cos(b) + a[0] * np.cos(b) + a[1] * np.cos(b * a[2]) + a[3] * np.cos(b * a[4]);

fig, ax = plt.subplots()
ax.plot(x, y)
```
